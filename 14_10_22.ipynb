```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def load_img(path):
  img = cv2.imread(path)#BGR
  return img[...,::-1]

img = load_img('Cubo (1).jpg') 

print(f'ALto: {img.shape[0]}')
print(f'Ancho: {img.shape[1]}')
print(f'Canales: {img.shape[2]}')
```

```python
fig, axs= plt.subplots(1,5, figsize=(20,5))
axs[0].imshow(img)
axs[0].set_title('Full picture')

hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
mask_r=cv2.inRange(hsv_img,np.array([0,0,250]),np.array([180,255,255]))
kernel = np.ones((7,7),np.uint8)

mask_r = cv2.morphologyEx(mask_r, cv2.MORPH_CLOSE, kernel)
mask_r = cv2.morphologyEx(mask_r, cv2.MORPH_OPEN, kernel)

seg_img = cv2.bitwise_and(img, img, mask=mask_r)
axs[1].imshow(seg_img)
axs[1].set_title('Orange')
###
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
mask_r=cv2.inRange(hsv_img,np.array([0,100,100]),np.array([180,255,200]))
kernel = np.ones((7,7),np.uint8)

mask_r = cv2.morphologyEx(mask_r, cv2.MORPH_CLOSE, kernel)
mask_r = cv2.morphologyEx(mask_r, cv2.MORPH_OPEN, kernel)

seg_img = cv2.bitwise_and(img, img, mask=mask_r)
axs[2].imshow(seg_img)
axs[2].set_title('Red')
###
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
mask_r=cv2.inRange(hsv_img,np.array([30,30,30]),np.array([90,255,255]))
kernel = np.ones((7,7),np.uint8)

mask_r = cv2.morphologyEx(mask_r, cv2.MORPH_CLOSE, kernel)
mask_r = cv2.morphologyEx(mask_r, cv2.MORPH_OPEN, kernel)

seg_img = cv2.bitwise_and(img, img, mask=mask_r)
axs[3].imshow(seg_img)
axs[3].set_title('Yellow')
##
hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
mask_r=cv2.inRange(hsv_img,np.array([0,0,0]),np.array([180,50,255]))
kernel = np.ones((7,7),np.uint8)

mask_r = cv2.morphologyEx(mask_r, cv2.MORPH_CLOSE, kernel)
mask_r = cv2.morphologyEx(mask_r, cv2.MORPH_OPEN, kernel)

seg_img = cv2.bitwise_and(img, img, mask=mask_r)
axs[4].imshow(seg_img)
axs[4].set_title('White')
```
